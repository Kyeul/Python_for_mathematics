```
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import math
```

```
#1.
#1-1

x = np.arange(-5, 10, 1.0)
y = 3**x
plt.axvline(x=0, color = 'black')
plt.axhline(y=0, color = 'black')
plt.plot(x, y)
plt.grid(True)
plt.show
```

```
#1.
#1-2
x = np.arange(0.1,5,0.1)
y = []
for i in x:
  y.append(math.log(i, 1/2))
plt.axvline(x=0, color = 'black')
plt.axhline(y=0, color = 'black')
plt.plot(x, y)
plt.grid(True)
plt.show
```
